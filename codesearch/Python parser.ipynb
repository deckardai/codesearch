```python
from pypy.interpreter.pyparser import pyparse
from pypy.interpreter.astcompiler import ast
from pypy.interpreter.astcompiler.astbuilder import ast_from_node
from pypy.tool.pytest.objspace import TinyObjSpace

from collections import defaultdict
```

```python
# Source code to tree (AST)
space = TinyObjSpace()
pyparser = pyparse.PythonParser(space)
compile_info = pyparse.CompileInfo("<filename>", "exec")


def source_to_tree(source):
    parsetree = pyparser.parse_source(source, compile_info)
    tree = ast_from_node(space, parsetree, compile_info)
    return tree
```

```python
# Tokenizer
import re
RE_TOKEN_SEP = re.compile(r'[^\w-]+')
RE_TOKEN_DEL = re.compile(r'[_-]+')


def normalize_name(name):
    return RE_TOKEN_DEL.sub('', name).lower()


def tokenize(name):
    return [normalize_name(token) for token in RE_TOKEN_SEP.split(name)]
```

```python
# Find nodes that define new names
def make_def_finders():
    # Direct reference by module name (node.names)
    # XXX Should use alias instead, for line numbers, support, with .. as, etc
    def Import(node):
        for name in node.names:
            # The alias or the root module goes into the namespace
            yield name.asname or name.name.split('.', 1)[0]

    ImportFrom = Import

    # Class Definition
    def ClassDef(node):
        yield node.name

    # Function Definition
    def FunctionDef(node):
        yield node.name

    # Variable name (node.id)
    def Name(node):
        if node.ctx == ast.Store or node.ctx == ast.AugStore or node.ctx == ast.Param:
            yield node.id
    
    ''' Attributes don't really define names
    # Attribute name (node.attr)
    def Attribute(node):
        if node.ctx == ast.Store or node.ctx == ast.AugStore:
            yield node.attr
    '''

    return locals()  # DRY

def_finders = make_def_finders()


# Find references to names
def make_ref_finders():
    # Direct reference by variable name (node.id)
    def Name(node):
        if node.ctx == ast.Load or node.ctx == ast.AugLoad:
            yield node.id

    ''' Attributes don't really use names
    # Direct reference by attribute name (node.attr)
    def Attribute(node):
        if node.ctx == ast.Load or node.ctx == ast.AugLoad:
            yield node.attr
    '''

    ''' Import don't really use names (they mention them)
    # Direct reference by module name (node.names)
    def Import(node):
        for alias in node.names:
            yield alias.name

    # Direct reference by module name (node.names)
    def ImportFrom(node):
        yield node.module
        for alias in node.names:
            yield alias.name
    '''
    return locals()  # DRY

ref_finders = make_ref_finders()


# Find constants
def constNum(node):
    yield node.n

def constStr(node):
    yield node.s


# Caracteristics of nodes
def get_type(node):
    return type(node).__name__

def is_scope(node):
    return get_type(node) in ['Module', 'ClassDef', 'FunctionDef']

def is_stmt(node):
    return isinstance(node, ast.stmt)
```

```python
''' Turn the tree into a data model
'''

DATA_FIELDS = [
    'id', 'name', 'attr', 'asname', 'module', 'level', 'op', 'ops', 'ctx',
    'nl', 'n', 's', 'lineno', 'col_offset',
]
LINK_FIELDS = [
    'value', 'target', 'test', 'iter', 'slice', 'dest', 'msg',
    'func',
    'globals', 'locals',
    'key', 'value', 'elt', 'arg',
    'left', 'right', 'operand',
    'context_expr', 'optional_vars',
    'type', 'inst', 'tback',
    'lower', 'upper', 'step',
    # Duplicates:
    'body', 'name', 'args',
]
LINK_SEQ_FIELDS = [
    'values', 'names', 'bases', 'body', 'targets',
    'keys', 'values', 'elts', 'generators', 'comparators',
    'orelse', 'decorator_list',
    'handlers', 'finalbody',
    'args', 'vararg', 'keywords', 'starargs', 'kwargs', 'defaults',
    'dims', 'ifs',
]
# XXX In the top-level class Expression (instead of Module), body is a simple
# reference -_- Also exec.body, call.args, excepthandler.name. Also
# Compare.ops is a list of numbers

# Since the naming is inconsistent, guess the kind of field at runtime
ALL_FIELDS = set(DATA_FIELDS + LINK_FIELDS + LINK_SEQ_FIELDS)

""" ctx meaning
Load = 1
Store = 2
Del = 3
AugLoad = 4
AugStore = 5
Param = 6

op meaning: astbuilder.augassign_operator_map
"""


def node_to_data(node):
    " Extract type, data, link types and links from a node "
    assert isinstance(node, ast.AST), 'Not an AST node'
    ntype = get_type(node)
    data = {'_type': ntype}
    links = {}

    for field in ALL_FIELDS:
        value = getattr(node, field, None)
        if value:
            if isinstance(value, ast.AST):
                links[field] = [value]
            elif isinstance(value, list) and isinstance(value[0], ast.AST):
                links[field] = value
            else:
                data[field] = value

    def_find = def_finders.get(ntype)
    if def_find:
        defines = map(normalize_name, def_find(node))
        if defines:
            data['defines'] = defines

    ref_find = ref_finders.get(ntype)
    if ref_find:
        uses = map(normalize_name, ref_find(node))
        if uses:
            data['uses'] = uses

    if is_scope(node):
        data['scope'] = True

    if is_stmt(node):
        data['stmt'] = True

    return data, links


def tree_to_data(tree):
    ''' Convert the tree into a data tree:
            node = (
                {field: data},
                {link_type: [node, ]}
            )
    '''
    data, obj_links = node_to_data(tree)
    links = {
        link_type: [
            tree_to_data(node)
            for node in nodes
        ]
        for link_type, nodes in obj_links.iteritems()
    }
    return data, links


def flatten_datatree(data_tree):
    ''' Returns a list of nodes data and a list of relationships:
        [node data, ], [(parent id, rel type, child id), ]
        Ids are indexes in the nodes list. 'rel type' is a string.
    '''
    nodes = []
    rels = []

    def new_node(data):
        nodes.append(data)
        return len(nodes) - 1

    def new_rel(this_id, link_type, child_id):
        rels.append((this_id, link_type, child_id))
        return len(rels) - 1

    def walk(tree):
        data, links = tree
        this_id = new_node(data)

        for link_type, children in links.iteritems():
            for child in children:
                child_id = walk(child)
                new_rel(this_id, link_type, child_id)

        return this_id
    walk(data_tree)
    return nodes, rels
```

```python
from py2neo import neo4j, node, rel

db = neo4j.GraphDatabaseService("http://localhost:7474/db/data/")

# Lucene index: idx = db.get_or_create_index(neo4j.Node, 'names')

# Label index
for prop in ['defines', 'uses', 'attr']:
    try:
        db.schema.create_index('code', prop)
    except ValueError as e:
        print str(e)

def ex_query(q):
    return neo4j.CypherQuery(db, q).execute()
```

```python
def red(s):
    return "\x1b[31m%s\x1b[0m" % s
def green(s):
    return "\x1b[32m%s\x1b[0m" % s
def blue(s):
    return "\x1b[34m%s\x1b[0m" % s
def magenta(s):
    return "\x1b[35m%s\x1b[0m" % s
def cyan(s):
    return "\x1b[36m%s\x1b[0m" % s
def bold(s):
    return "\x1b[1m%s\x1b[0m" % s


def print_neo(lines, n):
    if isinstance(n, list):
        for e in n:
            print blue(str(e)),
        print
    else:
        print blue(str(n))

    try:
        print '| %s' % magenta(lines[n['lineno'] - 1])
    except:
        pass


def print_neo_rel(lines, n):
    print_neo(lines, n)
    try:
        print '->', blue(' '.join([str((r.type, str(r.end_node))) for r in n.match_outgoing()]))
        print '<-', blue(' '.join([str((r.type, str(r.start_node))) for r in n.match_incoming()]))
    except:
        pass


def print_path(lines, p):
    for r in p:
        print_neo(lines, r.start_node)
        print r.type
    print_neo(lines, r.end_node)


def print_results(lines, res):
    print red('  - %s results' % len(res))
    for d in res.data:
        for i, col in enumerate(d.columns):
            n = d.values[i]
            print green(col), '=',
            print_neo(lines, n)
        print


def go_query(lines, q):
    res = ex_query(q)
    print_results(lines, res)
    return res
```

```python
# Get the relevant source code

def get_ctx(nodes, file_len, size=1):
    " Get line numbers surrounding the nodes"
    nos = set()
    for n in nodes:
        no = n['lineno']
        if no:
            for i in range(no - size, no + size + 1):
                nos.add(max(1, min(file_len, i)) - 1)
    return sorted(nos)


def get_lines(lines, nos):
    ' Make a text snippet from line numbers '
    return '\n'.join(map(lines.__getitem__, nos))


def highlight(text, word, color=magenta):
    return re.sub(r'\b%s\b' % word, color(word), text)
```

```python
# Parse source files and insert the trees into the db
# XXX Standardize properties and relationships names
# XXX Simplify, only care about outer statements. Include all the names used and defined by it, and lines range.
# XXX How to find parent of attributes then?

db.clear()

source_files = ['example.py', 'example_django.py']  # , '../pypy/pypy/interpreter/function.py']
file_lines = {}


for source_file in source_files:
    with open(source_file) as fd:
        source = fd.read()
    
    file_lines[source_file] = source.splitlines()
    
    tree = source_to_tree(source)

    data_tree = tree_to_data(tree)
    nodes, rels = flatten_datatree(data_tree)
    ret = db.create(*(
        map(node, nodes)
        +
        map(rel, rels)
    ))
    # Label the nodes
    batch = neo4j.WriteBatch(db)
    for n in ret:
        if isinstance(n, neo4j.Node):
            batch.add_labels(n, 'code')
    batch.submit()

    root = ret[0]
    db.create(node({'filename': source_file}), rel(0, 'CONTAINS_CODE', root))
```

```python
# Find uses and definitions of named entities. Scope-aware
# XXX Make a union, names or attr

q_path_scope = '''
match (use:code) where "%s" in use.uses
match (def:code) where "%s" in def.defines
match (file)-[*]->(use) where has(file.filename)
match path =
    (use) <-[*]- (scope {scope: True}) -[rdefs *0..6]-> (def)
    where none(rdef in rdefs where has(endNode(rdef).scope) and endNode(rdef) <> def)
return file, def, use, scope, path
'''


def find_path_def_use(name):
    norm = normalize_name(name)
    return ex_query(q_path_scope % (norm, norm))

# Find the definition of a name in a parent scope of a node (by id)
q_node_def = '''
start use=node(%i)
match path =
    (use) <-[*..20]- (scope {scope: True}) -[rdefs *0..6]-> (def)
    where "%s" in def.defines
    // Rewrite this with shortestPath?
    and none(rdef in rdefs where has(endNode(rdef).scope) and endNode(rdef) <> def)
return def, scope, path
'''


def find_def_by_id(nid, name):
    norm = normalize_name(name)
    return ex_query(q_node_def % (nid, norm))
```

```python
# Find named attributes (or key in a container). Dependency: their parent or container.
q_attr = '''
match path = (use:code {attr: "%s"}) --> (def:code)
match (file)-[*]->(use) where has(file.filename)
return file, use, def, path
'''


def find_attr(attr):
    return ex_query(q_attr % (attr))

# Find external names (import). Dependency: file or package where it is defined.
# TODO
```

```python
# Find friends (named entities in the same statement as a node). Stop at the statement level.
q_friends = '''
start center=node(%i)
match path = (center)-[rels *1..%i]-(n)
where has(n.id) and n.ctx = 1 and none(r in rels where has(startNode(r).stmt))
return n, n.id, length(path) as len
'''


def find_friends_by_id(nid, depth=10):
    return ex_query(q_friends % (nid, depth))
```

```python
# Find named products of an expression (definition of names by the parent statement):
# - target name
# - function that returns it or use it as default value
# - class that contains it or derives from it
# Getting parent statements that define names (direct or tuple assignement)
# XXX A bit specific with the return exception
q_products = '''
start n=node(%i)
match path =
    (n) <-[ups *0..%i]- (stmt {stmt: true}) -[downs *0..2]-> (def)
    where has(def.defines)
    and none(r in ups where has(endNode(r).stmt) and endNode(r)._type <> "Return")  // Traverse Return statements
    and none(r in downs where has(endNode(r).stmt))  // Maybe unnecessary
return def, stmt, path
'''


def find_products_by_id(nid, depth=3):
    return ex_query(q_products % (nid, depth))
```

```python
# Search by name, find other members of the same expression, all definitions, and the direct product
# XXX Should care more about external names and attributes on them

def search_deep(name, ctx_lines=0, verbose=False):
    print 'Looking for uses of "%s"' % blue(name)
    res_name = find_path_def_use(name)
    res_attr = find_attr(name)
    res_both = list(res_name) + list(res_attr)
    #if verbose: print_results(res)
    print red('- results: %s names, %s attributes' % (len(res_name), len(res_attr)))
    
    for row in res_both:
        filename = row['file']['filename']
        lines = file_lines[filename]
        use = row['use']

        print
        print '======== %s ========' % red(filename)
        if verbose: print_neo(lines, use)

        all_nodes = set(row['path'].nodes)
        
        if verbose: print
        if verbose: print 'Looking for close-by names'
        friends = find_friends_by_id(use._id)
        if verbose: print_results(lines, friends)
        
        if verbose: print
        if verbose: print 'Looking for definitions of these names'
        for f in friends:
            fn = f[0]
            fname = f[1]
            all_nodes.add(fn)  # In case no path is found
        
            if verbose: print '  - %s?' % blue(fname)
            friend_defs = find_def_by_id(fn._id, fname)
            for fdef in friend_defs:
                all_nodes.update(fdef['path'].nodes)
            if verbose: print_results(lines, friend_defs)
        
        if verbose: print
        if verbose: print 'Looking for products'
        products = find_products_by_id(use._id)
        if verbose: print_results(lines, products)
        
        print
        print red('---- Whole code sample ----')
        print highlight(get_lines(lines, get_ctx(
            all_nodes, file_len=len(lines), size=ctx_lines)), name)
        # XXX The context is missing the not-named expressions ("friends" are only the close named nodes)
    return res_both


'''search_deep('raw_proxy_view')
print
search_deep('path')
print
search_deep('fd')'''
print
res = search_deep('some_lib_call', ctx_lines=0, verbose=False)
```

```python
# Search by name, then by simple proximity

q_around = '''
start center=node(%i)
match (center)-[*0..%i]-(n)
return n
'''


def get_neighbors(node, depth=1):
    return [row[0] for row in ex_query(q_around % (node._id, depth))]


q_parents_children = '''
start center=node(%i)
match (p)-[*0..%i]->(center)-[*0..%i]->(c)
return p, c
'''

def get_lineage(node, depth=1):
    " Return parents and children "
    " XXX Returns duplicates "
    lineage = []
    for row in ex_query(q_parents_children % (node._id, depth, depth)):
        if row[0]:
            lineage.append(row[0])
        if row[1]:
            lineage.append(row[1])
    return lineage


# Search for a name, its definition, its neighbors, and print results
def search(name, ctx_lines=1):
    res = find_path_def_use(name)
    print red('%s results for "%s"' % (len(res), name))
    for i, row in enumerate(res):
        filename = row['file']['filename']
        lines = file_lines[filename]
        defi = row['def']
        use = row['use']
        path = row['path']
        around_def = get_neighbors(defi, depth=1) if defi else []
        around_use = get_neighbors(use, depth=2) if use else []
        path_nodes = path.nodes if path else []
        #around_path = reduce(list.__add__, map(get_neighbors, path_nodes, [1] * len(path_nodes)))

        print '\n%s) %s' % (i + 1, use)
        print red('---- Definition context ----')
        print highlight(get_lines(lines, get_ctx(
            around_def, file_len=len(lines), size=ctx_lines)), name)
        print
        print red('---- Use context ----')
        print highlight(get_lines(lines, get_ctx(
            around_use, file_len=len(lines), size=ctx_lines)), name)
        print
        print red('---- Path context ----')
        print highlight(get_lines(lines, get_ctx(
            path_nodes, file_len=len(lines), size=ctx_lines)), name)
    return res
```
